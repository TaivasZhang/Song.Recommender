```python
import pandas as pd

df = pd.read_csv('df_gui_1.csv') #dummy_data_gui.xlsx
```

```python
import numpy as np
df['hotness']= np.where(df['song_year_y'] == 2010, df['song_hotness_y'],df['song_hotness_x'])
df.sort_values(by=['hotness'], ascending=False).head(10)
```

```python
def return_song():
    input_string = input('Enter Song Title:')  
    indexer=df[df['song_title_x']==input_string].index
    cluster_num = df.loc[indexer,'prediction']
    cluster_list = df[df['prediction']==cluster_num.values.item(0)].sort_values(by=['hotness'], ascending=False).head(10)
    print(cluster_list['song_title_x'])
    
return_song()
```

## GUI Interface

```python
from tkinter import *

window = Tk()

window.title("Song Recommonder")
window.geometry('500x500')

lbl = Label(window, text="    Welcome to Nostalgic Songs    ", font=("Gilbert Color Bold", 36))

lbl.grid(column=1, row=0)

txt = Entry(window,width=30)
txt.grid(column=1, row=1, pady=10)

songs = ("         All Falls Down",\
         "         We Don't Care",\
         "   Don't Stop the Music",\
         "          Take A Bow",\
         "        Dead And Gone",\
         "           Sweet Lies",\
         "      Marshall Mathers",\
         "            Love Game")

list1 = StringVar(value = songs)

def clicked():
    lbl1 = Label(window, text="   Your Playlist   ", font=("Gilbert Bold", 20))
    lbl1.grid(column=1, row=3, pady=20)
    listbox1 = Listbox(window, height=len(songs), listvariable=list1, selectmode="browse", font=("Felt Tip Roman", 18))
    listbox1.grid(column=1, row=4)


btn = Button(window, text = 'Explore', font=("Gilbert Color Bold", 24), command = clicked)
btn.grid(column=1, row=2)

window.mainloop()
```
